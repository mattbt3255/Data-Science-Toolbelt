```python
# Halo 5 Public API (https://developer.haloapi.com/)
import requests
import json

API_KEY = 'INSERT KEY HERE' # Get Developer API Key at https://developer.haloapi.com/developer

headers = {'Ocp-Apim-Subscription-Key': API_KEY}

parameters = {'gamertag':'madballa55',
              'modes': 'arena',
              'start': '0',
              'count' : '25'}

api_url = 'https://www.haloapi.com/stats/h5/players/{gamertag}/matches?modes={modes}&start={start}&count={count}'.format(**parameters)

json_response = requests.get(api_url,headers=headers)

dicty = json.loads(json_response.text)

dicty
```
