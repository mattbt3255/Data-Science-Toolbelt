# Seaborn

```python
# EXAMPLES OF GRAPHS: http://seaborn.pydata.org/examples/index.html
# LIST OF FUNCTIONS: http://seaborn.pydata.org/api.html
# Datacamp: https://campus.datacamp.com/courses/data-visualization-with-seaborn
```

```python
# Scatterplot
import seaborn as sns

df = sns.load_dataset('tips')

sns.jointplot(x='total_bill',y='tip',data=df,kind ='reg')  # Kind are optional (can use hex or kde)

plt.show()
```

```python
# Histogram (distplot)
import matplotlib.pyplot as plt
import seaborn as sns

df = sns.load_dataset('tips')

sns.distplot(df['total_bill'],bins=10,kde=True)

plt.show()
```

```python
# Regression Plot
import matplotlib.pyplot as plt
import seaborn as sns

df = sns.load_dataset('tips')

sns.lmplot(data=df,x='total_bill',y='tip',col='sex',hue='sex')

plt.show()
```

```python
# Pairplot
import matplotlib.pyplot as plt
import seaborn as sns

df = sns.load_dataset('tips')

sns.pairplot(df,hue='sex',palette='coolwarm')

plt.show()
```

```python
# Bar Graph
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

df = sns.load_dataset('tips')

sns.barplot(x='sex',y='total_bill',data=df,estimator=np.mean)

plt.show()
```

```python
# Violin Plot
import matplotlib.pyplot as plt
import seaborn as sns

df = sns.load_dataset('tips')

sns.violinplot(x='day',y='total_bill',data=df,hue='sex',split=True)  # The width is the distribution (wider=more points)

plt.show()
```

```python
# Heatmap
import matplotlib.pyplot as plt
import seaborn as sns

df = sns.load_dataset('tips')

sns.heatmap(df.corr(),annot=True,cmap='coolwarm')

plt.show()
```

```python
# Seaborn Styling
import matplotlib.pyplot as plt
import seaborn as sns

df = sns.load_dataset('tips')

plot = sns.distplot(df['total_bill'])

# Set overall style
sns.set_style('white')

# Create Title
plot.set_title('Total Bill Distribution')

# Set Labels
plot.set(xlabel='Total Bill',ylabel='Count')

# Gets rid of y axis marks
sns.despine(left=True) 

plt.show()
```

# Plotly Dash

```python
# Scatterplot
import numpy as np
import seaborn as sns
import plotly.offline as pyo
import plotly.graph_objs as go

df = sns.load_dataset('tips')

data = [go.Scatter(x=df['total_bill'],
                   y=df['tip'],
                   mode='markers',
                   marker=dict(
                               size=12,
                               color='rgb(51,204,153)',
                               symbol='pentagon',
                               line = {'width':2}))]

layout = go.Layout(title='Total Bill vs Tips',
                   xaxis={'title':'Total Bill'},
                   yaxis=dict(title='Tip'),
                   hovermode='closest')

fig = go.Figure(data=data,layout=layout)

pyo.plot(fig,filename='Scatter.html')
```

```python
# Line Chart
import numpy as np
import plotly.offline as pyo
import plotly.graph_objs as go

x_values = np.linspace(0,1,100)
y_values = np.random.randn(100)

trace0 = go.Scatter(x=x_values,y=y_values+5,mode='markers',name='markers')
trace1 = go.Scatter(x=x_values,y=y_values,mode='lines',name='lines')
trace2 = go.Scatter(x=x_values,y=y_values-5,mode='lines+markers',name='lines+markers')



data = [trace0,trace1,trace2]

layout = go.Layout(title='Line Charts')

fig = go.Figure(data=data,layout=layout)

pyo.plot(fig,filename='Line.html')
```

```python
# HTML Components
import dash
import dash_html_components as html

app = dash.Dash()

app.layout = html.Div(['This is the outermost div!',
                       html.Div('This is an inner div!',style={'color':'red','border':'2px red solid'}),
                       html.Div('Another inner div!',style={'color':'blue','border':'3px blue solid'})],
                      style={'color':'black','border':'2px black solid'})

if __name__ == '__main__':
    app.run_server()
```

```python
# Core Components (https://dash.plot.ly/dash-core-components)
import dash
import dash_core_components as dcc
import dash_html_components as html

app = dash.Dash()

dropdown = dcc.Dropdown(options=[{'label':'New York City','value':'NYC'},
                                 {'label':'San Francisco','value':'SF'}],
                        value='SF') # Default Value

slider = dcc.Slider(min=-10,max=10,step=0.5,marks={i:i for i in range(-10,10)},value=0)

radio = dcc.RadioItems(options=[{'label':'New York City','value':'NYC'},
                                 {'label':'San Francisco','value':'SF'}],
                       value='SF')

app.layout = html.Div([html.Label('Dropdown'),dropdown,html.Label('Slider'),slider,html.P(html.Label('Radio')),radio])

if __name__ == '__main__':
    app.run_server()
```

```python
# Markdown Text
import dash
import dash_core_components as dcc
import dash_html_components as html

app = dash.Dash()

markdown_text = '''
### Dash and Markdown

Dash apps can be written in Markdown.
Dash uses the [CommonMark](http://commonmark.org/) specification of Markdown.

Check out their [60 Second Markdown Tutorial](http://commonmark.org/help/)
if this is your first introduction to Markdown!

Markdown includes syntax for things like **bold text** and *italics*,
[links](http://commonmark.org/help), inline `code` snippets, lists,
quotes, and more.
'''

app.layout = html.Div([
    dcc.Markdown(children=markdown_text)
])

if __name__ == '__main__':
    app.run_server()
```
